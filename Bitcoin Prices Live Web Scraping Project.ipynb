```python
# PROJECT 1: AUTOMATED WEB SCRAPER 

# https://www.worldometers.info/world-population/population-by-country/

from bs4 import BeautifulSoup
import requests

url = 'https://www.worldometers.info/world-population/population-by-country/'
    
page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')

print(soup)
```

```python
table = soup.find('table')
```

```python
table
```

```python
world_columns = table.find_all('th')
```

```python
world_column_titles = [columns.text for columns in world_columns]
print(world_column_titles)
```

```python
import pandas as pd

df2 = pd.DataFrame(columns = world_column_titles)
df2
```

```python
table.find_all('tr')
```

```python
world_rows = table.find_all('tr')

for row in world_rows[1:]:
    row_data = row.find_all('td')
    individual_row_data = [data.text for data in row_data]
    
    length = len(df2)
    df2.loc[length] = individual_row_data
    
# print(individual_row_data)

df2
```

### Bringing it all together

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from datetime import datetime
import os
import time
```

```python
def automated_crypto_pull():

    url = 'https://coinmarketcap.com/currencies/bitcoin/'

    page = requests.get(url)

    soup = BeautifulSoup(page.text, 'html')

    crypto_name = soup.find ('span', class_ = 'sc-65e7f566-0 lsTl').text [:7]


    crypto_price = soup.find ('span', class_ = 'sc-65e7f566-0 WXGwg base-text').text


    final_crypto = crypto_price.replace('$','')

    dt = datetime.now()

    dict = {'Crypto_Name' : crypto_name,
           'Price': final_crypto, 
           'TimeStamp': dt}

    cryptodf = pd.DataFrame([dict])

    if os.path.exists(r'/Users/boluwatifefadele/MY PORTFOLIO/New Projects/Automated_Crypto_Pull.csv'):
        cryptodf.to_csv(r'/Users/boluwatifefadele/MY PORTFOLIO/New Projects/Automated_Crypto_Pull.csv', mode = 'a', header = False, index = False)
    else:
        cryptodf.to_csv(r'/Users/boluwatifefadele/MY PORTFOLIO/New Projects/Automated_Crypto_Pull.csv', index = False)
    print(cryptodf)
```

```python
# automating it
while True:
    automated_crypto_pull()
    time.sleep(10)
```
